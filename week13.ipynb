```python
# 실습1 : 생일 기온 시각화
# 자신 생일의 서울 지역 최고 기온과 최저 기온을 아래와 같이 시각화
import csv
import matplotlib.pyplot as plt

# 2011.01.01-2021.01.01
f = open('week13.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)
high_list = [] 
low_list = [] 
for row in data:
    if row[-1] != '' : 
        if (row[0].split('-')[1] == '12') and (row[0].split('-')[2] == '10'):
            high_list.append(float(row[4])) 
            low_list.append(float(row[3]))      
plt.plot(high_list, color='pink')
plt.plot(low_list, color='skyblue')
plt.show()
```

```python
# 실습2 : 최고 기온 시각화
# 서울 지역의 1월과 8월 최고 기온 데이터를 아래와 같이 히스토그램으로 시각화
import csv
import matplotlib.pyplot as plt

# 2011.01.01-2021.01.01
f = open('week13.csv', 'r', encoding='cp949')
data = csv.reader(f)
next(data)
jan_list = [] 
aug_list = [] 
for row in data:
    if row[-1] != '' : 
        if(row[0].split('-')[1] == '01'): 
            jan_list.append(float(row[4])) 
        elif(row[0].split('-')[1] == '08'): 
            aug_list.append(float(row[4]))
plt.hist(aug_list, bins=100, color='red', label='August')
plt.hist(jan_list, bins=100, color='blue', label='January')
plt.legend() 
plt.show()
```
