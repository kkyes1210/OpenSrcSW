```python
# :  (  /        )

import csv
import numpy as np
import matplotlib.pyplot as plt

f = open('week14.csv', 'r', encoding='cp949')
d = csv.reader(f)
head = next(d)
head = next(d) 
data =[]
for row in d:
    temp = row[4:]
    temp = [int(x.replace(',', '')) for x in temp]
    temp.insert(0, row[1])
    data.append(temp)
f.close()

df= data[4] 
max = 0
min= 999999999999999999
max_list = [] 
min_list = [] 
total = 0

for i in range(0, len(data)):
    temp = data[i]
    if(temp[0] != '  '):
        total = 0
        for j in range(1, len(temp)):
            total = total + (int(df[j]) - int(temp[j])) ** 2
        if  total < min:
            min = total
            min_list = temp
        if  total > max:
            max = total
            max_list = temp

plt.figure(figsize=(9, 6))
plt.rc('font', family='Malgun Gothic') 
plt.style.use('ggplot')
plt.plot(df[1:], label="{0} /  : {1}".format(df[0], np.std(df[1:])))
plt.plot(max_list[1:], label="{0} /  : {1}".format(max_list[0], np.std(max_list[1:])))
plt.plot(min_list[1:], label="{0} /  : {1}".format(min_list[0], np.std(min_list[1:])))
plt.legend()
plt.show()
```
